```python
movie = input("Enter movie : ")
movies = get_movies_by_title(movie)
print("---------------MOVIES---------------")
for m in movies:
    print(m)

#will be done by select movie tool
movie:MovieModel = movies[0]

print("---------------THEATRES---------------")
theatres = get_theatres_by_movie_id(movie.id)
for t in theatres:
    print(t)

# will be done by select theatre tool
theatre = input("Select theatre : ")
theatre:TheatreModel = get_theatre_by_name(theatre)

print("---------------SHOWS---------------")
shows = get_shows_for_movie_with_theatre(movie_id=movie.id,theatre_id=theatre.id)
for s in shows:
    print(s)

# will be done by select show tool
show:ShowModel = shows[0]

print("---------------SEATS---------------")
seats = get_seats_for_show(show.id)
for s in seats:
    print(f"{s.seat_id}:{s.row_name}{s.seat_no}{"❌" if s.is_booked else "✅"}")

# rows = list(set([s.row_name for s in seats]))
# rows.sort()
# print("Screen this side")
# for row in rows:
#     if row=="C":
#         print("",end="  ")
#     for seat in seats:
#         if seat.row_name == row:
#             print(f"{seat.row_name}{seat.seat_no}",end=" ")
#     print()

# will be select seats tool
seats = int(input("Enter seats : "))

req:BookingRequest = BookingRequest(
    user_id=1,
    show_id=show.id,
    seat_ids=[seats],
    payment_method="UPI"
)
print(req)
booking_resp = book_tickets(req)
print(booking_resp)

#use get_ticket afterwards
```
